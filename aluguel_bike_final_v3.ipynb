## Q1: Influência do clima na quantidade de bikes alugadas

#### Média da quantidade de bicicletas alugadas quando o clima está ruim
Clima ruim: Pouca neve ou chuva / Tempestades / Nuvens dispersas e pouca chuva

```python
cnt_bad_weather = data[data['weathersit'] == 3][['cnt','dteday']]
cnt_bad_weather.mean()
```

#### Média da quantidade de bicicletas alugadas quando o clima está bom
Clima bom: Tempo limpo / Poucas nuvens / Parcialmente nublado

```python
cnt_good_weather = data[data['weathersit'] == 1][['cnt','dteday']]
cnt_good_weather.mean()
```

```python
pct = (cnt_good_weather.mean() / cnt_bad_weather.mean()) - 1
pct
```

### A média de bicicletas alugadas quando o clima está bom é 153% maior do que quando o clima está ruim.

## Q3: Diferença da quantidade de bikes alugadas nos anos de 2011 e 2012

#### Quantidade total de bikes alugadas nos anos de 2011 e 2012, respectivamente:

```python
total_ano1 = data[data['yr'] == 0]['cnt'].sum()
total_ano1
```

```python
total_ano2 = data[data['yr'] == 1]['cnt'].sum()
total_ano2
```

#### Cálculo da diferença do número de bicicletas alugadas anualmente (2011 e 2012)

```python
diff = total_ano2 - total_ano1
diff
```

### Houve um aumento de aproximadamente 153 milhões de bicicletas alugadas em 2012 (comparado ao ano de 2011)
